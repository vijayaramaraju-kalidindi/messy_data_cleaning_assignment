# Task 1 — Clean the Data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
print(f"Original DataFrame: \n{df}")
print(f"DataFrame Info: \n{df.info()}")
```

a. Impute missing values in age with the median and in city with the mode

```python
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)

print(f"DataFrame after filling missing 'age' values: \n{df}")

city_mode = df['city'].mode()[0]
df['city'] = df['city'].fillna(city_mode)

print(f"DataFrame after filling missing 'city' values: \n{df}")
```

b. Remove any duplicate rows

```python
duplicate_rows = df[df.duplicated()]
print(f"\nDuplicate rows: \n{duplicate_rows}")

df = df.drop_duplicates()
print(f"\nDataFrame after removing duplicates: \n{df}")
```

# Task 2 — Encode Categorical Columns

a. Apply One-Hot Encoding to city

```python
df = pd.get_dummies(df, columns=['city'])
print(f"\nDataFrame after encoding 'city' column: \n{df}")
```

b. Apply Label Encoding to gender

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
print(f"\nDataFrame after encoding 'gender' column: \n{df}")
```

# Task 3 — Scale Numerical Features
Scale age and annual_income using both methods below on the cleaned dataframe, and print the results of each:

a. Min-Max Scaling (using MinMaxScaler)

```python
df_minmax = df.copy()
for col in ['age', 'annual_income']:
    df_minmax[col] = (df_minmax[col] - df_minmax[col].min()) / (df_minmax[col].max() - df_minmax[col].min())
print(f"\nDataFrame after Min-Max scaling: \n{df_minmax}")
```

b. Standardisation (using StandardScaler)

```python
df_std = df.copy()
for col in ['age', 'annual_income']:
    df_std[col] = (df_std[col] - df_std[col].mean()) / df_std[col].std()
print(f"\nDataFrame after Standardization: \n{df_std}")
```

Min-Max Scaling

Use when you need values in a fixed range (0–1)

Standardization

Use when data has outliers or unknown distribution
